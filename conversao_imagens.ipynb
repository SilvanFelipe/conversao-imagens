```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

# Função para converter imagem colorida para níveis de cinza
def color_to_grayscale(image):
    # Fórmula de luminosidade: Gray = 0.299*R + 0.587*G + 0.114*B
    grayscale = np.zeros((image.shape[0], image.shape[1]), dtype=np.uint8)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            r, g, b = image[i, j]
            grayscale[i, j] = int(0.299 * r + 0.587 * g + 0.114 * b)
    return grayscale

# Função para converter imagem em níveis de cinza para binária
def grayscale_to_binary(image, threshold=127):
    binary = np.zeros_like(image, dtype=np.uint8)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            binary[i, j] = 255 if image[i, j] > threshold else 0
    return binary

# Carregando a imagem de entrada
input_image = cv2.imread('lenna_colorida.png')
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)

# Convertendo para níveis de cinza
grayscale_image = color_to_grayscale(input_image)

# Convertendo para imagem binária
binary_image = grayscale_to_binary(grayscale_image)

# Visualizando as imagens
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(input_image)
plt.title("Imagem Colorida")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(grayscale_image, cmap='gray')
plt.title("Imagem em Níveis de Cinza")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(binary_image, cmap='gray')
plt.title("Imagem Binarizada (P&B)")
plt.axis('off')

plt.tight_layout()
plt.show()
```
